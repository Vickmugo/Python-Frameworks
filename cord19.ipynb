```python
# Remove rows without titles or abstracts
cord19_clean = cord19.dropna(subset=['title', 'abstract']).copy()
```

```python
# Fill missing journals with 'Unknown'
cord19_clean['journal'] = cord19_clean['journal'].fillna('Unknown')
```

```python
# Convert publish_time to datetime
cord19_clean['publish_time'] = pd.to_datetime(cord19_clean['publish_time'], errors='coerce')
```

```python
 #Extract publication year
cord19_clean['year'] = cord19_clean['publish_time'].dt.year
```

```python
# Create new column: abstract word count
cord19_clean['abstract_word_count'] = cord19_clean['abstract'].apply(lambda x: len(str(x).split()))
```

### Data Analysis and Visualisation

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# 1. Count papers by publication year
year_counts = cord19_clean['year'].value_counts().sort_index()
print(year_counts)
```

```python
plt.figure(figsize=(10,5))
year_counts.plot(kind='bar')
plt.title("Number of Publications by Year")
plt.xlabel("Year")
plt.ylabel("Paper Count")
plt.show()
```

```python
# 2. Top journals publishing COVID-19 research
top_journals = cord19_clean['journal'].value_counts().head(10)
print(top_journals)
```

```python
plt.figure(figsize=(10,5))
top_journals.plot(kind='bar')
plt.title("Top Publishing Journals")
plt.xlabel("Journal")
plt.ylabel("Paper Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Word frequency in titles (WordCloud)
all_titles = " ".join(cord19_clean['title'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_titles)
```

```python
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Frequent Words in Titles")
plt.show()
```

```python
# 4. Distribution of papers by source
source_counts = cord19_clean['source_x'].value_counts().head(10)
plt.figure(figsize=(10,5))
source_counts.plot(kind='bar')
plt.title("Distribution by Source")
plt.xlabel("Source")
plt.ylabel("Paper Count")
plt.show()
```

```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# Load data
@st.cache_data
def load_data():
    df = pd.read_csv("/content/drive/MyDrive/cord19.csv", low_memory=False)
    df = df.dropna(subset=['title', 'abstract']).copy()
    df['journal'] = df['journal'].fillna('Unknown')
    df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce')
    df['year'] = df['publish_time'].dt.year
    return df
```

```python
df = load_data()

st.title("CORD-19 Data Explorer")
st.write("Explore COVID-19 research papers interactively.")
```

```python
# Filter by year
min_year, max_year = int(df['year'].min()), int(df['year'].max())
year_range = st.slider("Select year range", min_year, max_year, (2020, 2021))
filtered = df[(df['year'] >= year_range[0]) & (df['year'] <= year_range[1])]

st.subheader("Sample Data")
st.write(filtered[['title', 'journal', 'year']].head())
```

```python
# Publications over time
st.subheader("Publications Over Time")
counts = filtered['year'].value_counts().sort_index()
fig, ax = plt.subplots()
counts.plot(kind='bar', ax=ax)
ax.set_title("Publications per Year")
st.pyplot(fig)
```

```python
# Top journals
st.subheader("Top Journals")
top_journals = filtered['journal'].value_counts().head(10)
fig, ax = plt.subplots()
top_journals.plot(kind='bar', ax=ax)
ax.set_title("Top Journals")
st.pyplot(fig)
```

```python
# Word Cloud
st.subheader("Word Cloud of Titles")
all_titles = " ".join(filtered['title'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_titles)
fig, ax = plt.subplots()
ax.imshow(wordcloud, interpolation='bilinear')
ax.axis("off")
st.pyplot(fig)
```
